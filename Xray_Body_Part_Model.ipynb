```python
#import dependencies
import pandas as pd
import numpy as np
```

```python
#importing csv
Image_Paths_df = pd.read_csv('Transformed_Valid_Data/Image_paths.csv')
Image_Paths_df.head()
```

```python
Seperated_Path_df = pd.read_csv('Transformed_Valid_Data/seperated_path.csv')
Seperated_Path_df.head()
```

```python
len(Seperated_Path_df)
```

```python
XRay_Type_df = Seperated_Path_df.drop(
    Seperated_Path_df.columns[[0,1,3,4,5]], axis=1)
XRay_Type_df.head()
```

```python
XRay_Type_df.dtypes
```

```python
Body_Part_Data_df = pd.merge(XRay_Type_df, Image_Paths_df, left_index=True, right_index=True)
Body_Part_Data_df.head()
```

```python
rows = len(Body_Part_Data_df)
small_df = Body_Part_Data_df.head(3150)
Body_Part_Data_df.astype('str')
small_df.astype('str')
```

```python
#from keras_preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator(rescale=1./255)
train_generator = datagen.flow_from_dataframe(dataframe=small_df, directory = "./",
                                             x_col='Image_Path', y_col='Xray_Type',
                                             class_mode='categorical', target_size=(256,256),
                                             color_mode='grayscale', batch_size=32)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Keras
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
from keras.datasets import mnist
```

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), padding='same',
                 input_shape=(256,256,1)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32,(3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3), padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(7))
model.add(Activation('sigmoid'))

model.compile(loss='categorical_crossentropy', 
              optimizer='adam',metrics=['accuracy'])

model.fit_generator(train_generator,
                    steps_per_epoch=3150//32,
                    shuffle=True,
                    epochs=25)
```

```python
model.save("body_part_model.h5")
```
